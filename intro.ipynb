## Weight setting

Weights are needed to produce valid population-level estimates. Several stages of nonresponse adjustment and poststratification are done to reduce sampling error and bias. 


For the 2001 NHTS, two sets of weights have been provided: 

- The full sample, including add-on samples for selected areas as well as the national sample; 
- The national sample only.


The public use dataset contains person and household weights for: 

- Useable households in which person interviews were completed with at least 50 percent of adults in the household.  
(69,817 households in the national and add-on samples, 26,038 households in the national sample) 

- 100 percent households in which person interviews were completed with all adults in the household.  
(60,520 households in the national and add-on samples, 22,178 households in the national sample)


### Replicate weights

**Replicate weights** - 每个ID对应的99个相似的权重值，目的是为了求解标准误差(standard errors)
 
**Replicate variance estimation** - The idea is that sample estimates are made for a number of subsamples of the fully conducted survey. One then looks at the difference between each "replicate sample estimate" and the "full sample estimate" and squares the difference. Finally, one sums up the squared differences across all the replicates, with an appropriate multiplicative factor.

The replicate weights were calculated using the [**delete-one Jackknife method**](https://en.wikipedia.org/wiki/Jackknife_resampling#CITEREFEfron1982). These weights can be used to calculate standard error estimates using WesVar or SUDAAN. **Standard error estimates** can also be easily calculated using the following formula:

$$\sqrt{\frac{98}{99} \sum_{i=1}^{99}[R E P(i)-x]^{2}}$$

where $x$ is the full sample estimate (calculated by using the full sample weights) and 
$REP(i)$ is the estimate calculated by using the replicate weights and the summation over the index $i$ is from 1 to 99. 

For example, suppose one is interested in an estimate of persons for Option 1 using the full sample. The weight **WTHHFIN** is used to calculate the overall estimate $x$. The weight **WTHHFIN1** is used to calculate the estimate $REP(1)$, the weight **WTHHFIN2** is used to calculate the estimate $REP(2)$, etc. 

Replicate weights are provided only for households and persons. For vehicles, use the household replicate weights. For travel day trips, use the person weight times 365. For travel period trips, use the person weight times 365/28.

## Transferability Statistics

[**2001 Transferability**](https://nhts.ornl.gov/tx/) refers to the transferability of the 2001 Nationwide Household Travel Survey (NHTS) results to the regional or local level. Transferability is a methodology for using the NHTS data to estimate regional or local travel, including:

1. **vehicle trips (VT), vehicle miles of travel (VMT), person trips (PT), person miles of travel (PMT)**
- A person trip(PT) is a trip by one or more persons in any mode of transportation. 
  Each person is considered as taking one person trip. 
- A person mile of travel(PMT) is a mile traveled by one or more persons.  
  A non-air person mile of travel(PMTNA) simply excludes air travel.  
- A vehicle trip(VT) is a trip by a single privately owned vehicle (POV) regardless of the number of persons in the vehicle.  
- A vehicle mile of travel(VMT) is a mile traveled in a private vehicle.  
For example, if a household of 4 persons all took a 5-mile trip together in a POV, that household will have taken 4 person trips, 20 person miles of travel (of all which are non-air), 1 vehicle trip, and 5 vehicle miles of travel.

2. **trip purposes** 
- Percent of Work Trips (Home-Based Work)
- Percent of Shopping Trips (Home-Based Shopping)
- Percent of Social-Recreational Trips (Home-Based Soc/Rec)
- Percent of Other Trips (Home-Based Other)
- Percent of non-Home-Based Trips

Trip generation rate information for selected **Census Tracts** and **Transportation Analysis Zones (TAZ)**.


### Transfer Method

参考文献：[Transferring 2001 National Household Travel Survey](https://nhts.ornl.gov/2001/pub/TransferabilityReport.pdf)

#### Census tract level

**聚类和回归**

1. 前提

- The majority of the households within a Census tract are sufficiently homogeneous with respect to the travel determinants (e.g., household income, employment status, life cycle, vehicle ownership).

- Census tracts can be grouped into homogeneous clusters with respect to travel determinants reported for each tract.

- There are enough NHTS households assigned to each cluster, and NHTS households in a cluster are representative of all households within that cluster, with respect to travel determinants.

2. 步骤

**Step1. Identification of Significant Travel Determinants**

household size, household vehicle count, the percent and number of workers in the Census tract who take public transit to work, the ACCRA cost of living and cost of transportation indices, population density, household income and buying power, Census division and region, MSA size, life cycle, and the number of bus and train routes in a Census tract

**Step2. Clustering of Census Tracts**

初始聚类标准：Tracts with a median household income, resulting in a total of 11 clusters.

经过附加区域的评估修改：The revised clustering scheme results in 10 clusters of Census tracts.

之后把相应的NHTS家庭分配到对应的聚类中。

**Step3. Assessments of Cluster-based Approach**

(i)Similar to the 1995 assessment, the superiority of alternative estimation methods is determined by the number of times a given method produces estimates that are closest to the baseline standards for all travel parameters (i.e., PT and VT).

(ii)Unlike the 1995 assessment, alternative estimation methods were assessed in 2001 on the basis of the mean absolute deviation from the baseline standards of all 9 add-on areas.

→但发现聚类方法产生的结果不太理想。

**Step4. Regression within Geo-economic Clusters**

The five geo-economic clusters: mega-urban, urban, suburban, rural and extreme-poverty

回归模型——针对每个出行参数(PT、PMT、VT、VMT)和地理经济聚类组合。（最小二乘法估计的线性模型）

1 Estimation of Trip Frequency

2 Assessment of Trip Frequency Estimates

3 Estimation of Household PMT and VMT

4 Assessment of PMT and VMT Estimates

5 Estimation of Trips by Trip Purpose (Logistic Regression)

6 Assessment of Trip Estimates by Trip Purpose

#### TAZ level

**Step1.** Travel estimates at the Census tract level were first disaggregated to the block centroid level.

**Step2.** The estimates at the block centroid level were then aggregated to the TAZ(Transportation Analysis Zones) level.

Census tract→Block→Part of TAZ

**Step3.**

→Estimated Average PT per Household(1), No. of Households(2)→Total PT=(1)*(2)

→Weighted average PT per household = Total PT / Total Household

## Trip chaining

解读Trip, Trip Chaining, Tour之间的相互转换。

**Direct Trip**: A trip that travels directly between two anchor destinations, such as a trip from home to work.  
**Chain**: A series of short trips linked together between anchor destinations, such as a trip that leaves home, stops to drop a passenger, stops for coffee, and continues to work.  
**Intervening Stop**: The stops associated with chained trips.  
**Tour**: The total travel between two anchor destinations, such as home and work, including **both direct trips and chained trips with intervening stops**.  

More details can be found in the "What Can Be Found On This CD.doc" .
